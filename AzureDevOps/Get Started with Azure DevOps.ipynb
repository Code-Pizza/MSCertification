## Value Stream Maps (VSMs)
The purpose of a VSM is to visually show where in the process a team creates value and where there's waste. <br>
The goal, of course, is to arrive at a process that delivers maximum value to the customer with minimum waste.

### Development processes
New features requested through My IT Requests or email <br>
The BI team then creates a User Story and tasks in Azure DevOps <br>
The story is then assigned points and ready for backlog grooming <br>
Once groomed, the story is assigned to a developer and sprint <br>
When the sprint arrives, the developer begins work and delivers by the due date <br>

### Test processes
A BI team member is notified that pull request is available <br>
The BI manager or team member completes a code review and provides feedback/approval <br>
The developer responds to the feedback and/or closes the request once approved <br>

### Operation Processes
The BI developer has to wait until the pull request is completed by the BI manager or another team member <br>
Then the developer can deploy the updated to UAT or Prod

### Calculating Customer Value Artifacts

- Total lead time:
    - The time for a feature to make it to the customer
- Process time:
    - The time spent on a feature that has value to the customer
 - Activity Ratio is the process time divided by the lead time <br>

    $  Activity Ratio = \frac{Process Time}{Lead Time} $

## DevOps Defined

DevOps is the union of people, process, and products to enable continuous delivery of value to our customers.<br>
Governed by a set of practices:
- Agile planning. 
    - Create a backlog of work that everyone on the team and in management can see
    - Prioritize the items so we know what we need to work on first
- Continuous integration (CI). 
    - Automate build and test processes. 
    - Run every time a team member commits changes to version control.
- Continuous delivery (CD). 
    - CD is how we test, configure, and deploy from a build to a QA or production environment.
- Monitoring
    - Use telemetry to get information about performance and usage patterns

## Elite Performing Teams

- Deploy more frequently - up to dozens of times per day.
- Reduce lead time from commit to deploy
- Reduce change failure rate
- Recover from incidents more quickly

## Delivery Plans Defined

A delivery plan is a visualization of one or more work schedules. <br>
It is intended to provide teams and management an overall view of what each team is planning to produce and when. <br>

- Questions to ask whe reviewing the plan:
    - Are we sure we can deliver what we have committed to on our current schedule?
    - Are we confident that the teams we depend on will deliver what we need on their current schedule?
    - Are there lulls in our schedule that we could fill with work?

## Recommedations

## Azure Delivery Plans

Delivery Plans is an extension for Azure DevOps that helps organizations plan and review work schedules across multiple teams. <br>
A visual representation of one or more work schedules against a calendar backdrop

- Use delivery plans to:
    - Gauge the likelihood of other teams producing their deliverables on their current schedules based on their commitments
    - Plan your team's work schedule at least in part based on the work schedules of other teams.

### Plan a Sprint Using Delivery Plans

- Create a delivery plan.
- Add team sprints and milestones.
- Rearrange work items to fit the overall schedule.
