# Group and bin data for analysis
- Grouping is used for categories of data.
- Binning is similar to grouping, but it is used for grouping continuous fields, such as numbers and dates.
- In a bar chart you can highlight multiple columns and then group them together based on the field values

# Create bin groups
The process of binning allows you to group your numerical and time field data into "bins" of equal size. This approach allows you to visualize and identify trends in your data in more meaningful ways. Binning allows you to right-size the data that Power BI Desktop displays.


# Apply clustering techniques
Clustering allows you to identify a segment (cluster) of data that is similar to each other but dissimilar to the rest of the data. The process of clustering is different to that of grouping, which you accomplished previously. 

Add a scatter plot and then direct it Automatically find clusters  
Define the number of clusters desired - and it will color by cluster color  

If you want to find clusters by using more than two measures, you can use a table visual instead, add all the fields you want to use, and then run the clustering algorithm by using the same process.

# Conduct time series analysis
Time series analysis involves analyzing a series of data in time order to identify meaningful information and trends and make predictions. The result of time series analysis is the best data that you can use for forecasting activities.

Use a visualization type that is suitable for displaying trends and changes over time, such as a line chart, area chart, or scatter chart. You can also import a time series custom visual into Power BI Desktop from Microsoft AppSource.  



# Use the Analyze feature
The Analyze feature provides you with additional analysis that is generated by Power BI for a selected data point. 

useful for analyzing why your data distribution looks the way that it does.



# Review Quick insights
The Quick insights feature in Power BI uses machine learning algorithms to go over your entire dataset and produce insights (results) for you quickly. This feature is a great way to build dashboards when you don't know where to start.  

Quick Insights page contains up to 32 separate insight cards, and each card has a chart or graph plus a short description.  

# Apply AI Insights
The AI Insights feature allows you to connect to a collection of pretrained machine learning models that you can apply to your data to enhance your data preparation efforts.  

Use the add column feature to ad an AI insights feature
